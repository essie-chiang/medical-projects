```python
from tool.config_utils import process_config
from data_process import LabelDataReader, DataReader

try:
    config = process_config('segmention_config2.json')
#    print(config)
except Exception as e:
    print('[Exception] Config Error, %s' % e)
    exit(0)
dr = LabelDataReader(config)
dr.init()
no_list, lower_list, full_list, label_list = dr.get_show_data()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np


for i in range(8):
    plt.figure(figsize=(15,5))
    plt.subplot(2,3,1)
    plt.title("no_" + str(i))
    no_img = no_list[i]
    print("no_img", no_img.max(), no_img.min(), no_img.mean())
    arr=no_img.flatten()
    plt.hist(arr, bins=15, normed=1,edgecolor='None',facecolor='red')
    plt.imshow(no_img, cmap='gray')
    
#    plt.subplot(2,3,2)
#    plt.title("low_" + str(i))
#    lower_img = lower_list[i]
#    print(lower_img.max(), lower_img.min(), lower_img.mean())
#    arr=lower_img.flatten()
#    plt.hist(arr, bins=15, normed=1,edgecolor='None',facecolor='red')
    
    plt.subplot(2,3,2)
    plt.title("full_" + str(i))
    full_img = full_list[i]
    print("full_img",full_img.max(), full_img.min(), full_img.mean())
    arr=full_img.flatten()
    plt.hist(arr, bins=15, normed=1,edgecolor='None',facecolor='red')
    plt.imshow(full_img, cmap='gray')
    
    subno = full_img - no_img
    sub_no_zero = subno # np.where(subno > 0, 0, subno)
    new_no_img = np.where(full_img > no_img, no_img, full_img)
    new_sub = full_img - new_no_img
    newnew_no = full_img - new_sub
    
    plt.subplot(2,3,3)
    plt.title("new_no_img" + str(i))
    print("new_no_img",new_no_img.max(), new_no_img.min(), new_no_img.mean())
    arr=new_no_img.flatten()
    plt.hist(arr, bins=15, normed=1,edgecolor='None',facecolor='red')
    plt.imshow(new_no_img, cmap='gray')
        
    plt.subplot(2,3,4)
    plt.title("new_sub" + str(i))
    print("new_sub", new_sub.max(), new_sub.min(), new_sub.mean())
    arr=new_sub.flatten()
    plt.hist(arr, bins=15, normed=1,edgecolor='None',facecolor='red')
    
    plt.subplot(2,3,5)
    plt.title("new_sub" + str(i))
    print("new_sub", new_sub.max(), new_sub.min(), new_sub.mean())
    arr=new_sub.flatten()
    plt.hist(arr, bins=15, normed=1,edgecolor='None',facecolor='red')
    plt.imshow(new_sub, cmap='gray')
    
    plt.subplot(2,3,6)
    plt.title("newnew_no" + str(i))
    print("newnew_no", newnew_no.max(), newnew_no.min(), newnew_no.mean())
    arr=newnew_no.flatten()
    plt.hist(arr, bins=15, normed=1,edgecolor='None',facecolor='red')
    plt.imshow(newnew_no, cmap='gray')
    
    plt.show()
```
