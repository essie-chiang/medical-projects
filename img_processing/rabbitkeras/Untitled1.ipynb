```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
full = cv2.imread('datasets/8/full-contrast.png', 0)
low = cv2.imread('datasets/8/lower-contrast.png', 0)
no = cv2.imread('datasets/8/no-contrast.png', 0)

full = np.array(full).astype('float32')
low = np.array(low).astype('float32')
no = np.array(no).astype('float32')

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.title("no_")
plt.imshow(no, cmap='gray')
plt.axis('off') 

plt.subplot(1,3,2)
plt.title("low_")
plt.imshow(low, cmap='gray')
plt.axis('off') 

plt.subplot(1,3,3)
plt.title("full_")
plt.imshow(full, cmap='gray')
plt.axis('off') 
```

```python
img_full_test = full.copy()

mask = list()
for bin in reversed(range(0, 2)):
    cbin = 128 * bin
    imgcb = np.where((full - cbin) >= 0, img_full_test, 0)
    mask.append(imgcb.copy())
    img_full_test -= imgcb

plt.figure(figsize=(16, 4))
for i in range(len(mask)):
    img = mask[i]
    print(img.mean())
    plt.subplot(1, len(mask), (i+1))
    plt.imshow(img, cmap='gray')
plt.show()

mask0 = mask[i]

plt.figure(figsize=(16, 4))
low = np.where(mask0>0, low, 0)
plt.subplot(1, 2, 1)
plt.imshow(low, cmap='gray')

no = np.where(mask0>0, no, 0)
plt.subplot(1, 2, 2)
plt.imshow(no, cmap='gray')

plt.show()
```
